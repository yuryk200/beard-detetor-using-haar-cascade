```python
#imports
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import os
from PIL import Image, ImageFilter
```

```python
#setting up camera to capture
cap = cv.VideoCapture(2)

#Capturing if a bool to see if there is an image, and the actual frame itself  
ret, frame = cap.read()



```

```python
#convert camera frame from bgr to rgb
frame_rgb = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
# show camera frame
plt.imshow(frame_rgb)
plt.title('Test frame')
plt.show()

# close camera
cap.release()
```

```python
# beard_cascade = cv.CascadeClassifier('cascade.xml')
beard_cascade = cv.CascadeClassifier('cascade(2).xml')

def beard(img):
    beard_img = img.copy()
    beard_rect = beard_cascade.detectMultiScale(beard_img,scaleFactor = 1.1, minNeighbors = 7)
    
    for (x,y,w,h) in beard_rect:
        cv.rectangle(beard_img, (x, y), (x + w, y + h), (0,0,0), 10)
    return beard_img
```

```python
img = frame
img_copy1 = img.copy()

beardframe = beard(img_copy1)
beardimg = cv.cvtColor(beardframe, cv.COLOR_BGR2RGB)
plt.imshow(beardimg)
```
